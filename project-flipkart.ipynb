## Convert XML → YOLO

```python
import os
import xml.etree.ElementTree as ET

OUTPUT_LABELS = "/kaggle/working/converted_dataset/plate/labels"
os.makedirs(OUTPUT_LABELS, exist_ok=True)

def convert_xml_to_yolo(xml_path, output_path):
    tree = ET.parse(xml_path)
    root = tree.getroot()

    width = int(root.find("size/width").text)
    height = int(root.find("size/height").text)

    with open(output_path, "w") as f:

        for obj in root.findall("object"):

            bbox = obj.find("bndbox")

            xmin = float(bbox.find("xmin").text)
            ymin = float(bbox.find("ymin").text)
            xmax = float(bbox.find("xmax").text)
            ymax = float(bbox.find("ymax").text)

            x_center = ((xmin + xmax) / 2) / width
            y_center = ((ymin + ymax) / 2) / height
            box_width = (xmax - xmin) / width
            box_height = (ymax - ymin) / height

            class_id = 0   # license_plate

            f.write(
                f"{class_id} "
                f"{x_center:.6f} "
                f"{y_center:.6f} "
                f"{box_width:.6f} "
                f"{box_height:.6f}\n"
            )

for file in os.listdir(PLATE["google_images"]):
    if file.endswith(".xml"):

        xml_path = os.path.join(PLATE["google_images"], file)

        txt_path = os.path.join(
            OUTPUT_LABELS,
            file.replace(".xml", ".txt")
        )

        convert_xml_to_yolo(xml_path, txt_path)

print("Google Images conversion completed!")
```

## Function to remap YOLO labels

```python
def remap_yolo_labels(input_file, output_file, class_map):

    with open(input_file, "r") as f:
        lines = f.readlines()

    new_lines = []

    for line in lines:

        parts = line.strip().split()

        if len(parts) == 0:
            continue

        old_class = int(parts[0])

        if old_class not in class_map:
            continue

        new_class = class_map[old_class]

        parts[0] = str(new_class)

        new_lines.append(" ".join(parts))

    with open(output_file, "w") as f:
        f.write("\n".join(new_lines))
```

## Standardize all training labels

```python
import os

input_dir = HELMET["train_labels"]
output_dir = os.path.join(STANDARDIZED, "helmet/labels/train")

count = 0

for file in os.listdir(input_dir):

    if file.endswith(".txt"):

        remap_yolo_labels(
            os.path.join(input_dir, file),
            os.path.join(output_dir, file),
            HELMET_MAP
        )

        count += 1

print("Converted:", count)
```

```python
import os
from collections import defaultdict

COCO_LABELS = os.path.join(STANDARDIZED, "coco/labels/train")

class_to_images = defaultdict(list)

for label_file in os.listdir(COCO_LABELS):

    if not label_file.endswith(".txt"):
        continue

    path = os.path.join(COCO_LABELS, label_file)

    classes = set()

    with open(path) as f:
        for line in f:

            if line.strip():

                cls = int(line.split()[0])

                classes.add(cls)

    for cls in classes:

        if cls in TARGET_COUNTS:

            class_to_images[cls].append(label_file)
```

```python
import random

random.seed(42)

selected_images = set()

for cls, target in TARGET_COUNTS.items():

    images = class_to_images[cls]

    if len(images) >= target:
        chosen = random.sample(images, target)
    else:
        chosen = images

    selected_images.update(chosen)

print("Total unique images:", len(selected_images))
```

```python
import os
import shutil

def merge_dataset(image_dir,
                  label_dir,
                  out_image_dir,
                  out_label_dir,
                  prefix):

    copied = 0

    for img in os.listdir(image_dir):

        if not img.lower().endswith((".jpg",".jpeg",".png")):
            continue

        base = os.path.splitext(img)[0]

        label = base + ".txt"

        src_img = os.path.join(image_dir, img)
        src_lbl = os.path.join(label_dir, label)

        new_img = f"{prefix}_{img}"
        new_lbl = f"{prefix}_{label}"

        shutil.copy2(src_img,
                     os.path.join(out_image_dir,new_img))

        if os.path.exists(src_lbl):

            shutil.copy2(src_lbl,
                         os.path.join(out_label_dir,new_lbl))

        copied += 1

    print(f"{prefix}: {copied} images merged.")
```

```python
import os

def remap_yolo_labels(input_file, output_file, class_map):

    new_lines = []

    with open(input_file, "r") as f:

        for line in f:

            parts = line.strip().split()

            if len(parts) == 0:
                continue

            cls = int(parts[0])

            # Ignore unwanted classes
            if cls not in class_map:
                continue

            parts[0] = str(class_map[cls])

            new_lines.append(" ".join(parts))

    if len(new_lines):

        with open(output_file, "w") as f:
            f.write("\n".join(new_lines))
```

```python
import os
import shutil
import random

train_img = "/kaggle/working/model1_dataset/images/train"
train_lbl = "/kaggle/working/model1_dataset/labels/train"

val_img = "/kaggle/working/model1_dataset/images/val"
val_lbl = "/kaggle/working/model1_dataset/labels/val"

os.makedirs(val_img, exist_ok=True)
os.makedirs(val_lbl, exist_ok=True)

images = os.listdir(train_img)

random.seed(42)

val_images = random.sample(images, int(0.1 * len(images)))

for img in val_images:

    shutil.move(
        os.path.join(train_img, img),
        os.path.join(val_img, img)
    )

    label = os.path.splitext(img)[0] + ".txt"

    shutil.move(
        os.path.join(train_lbl, label),
        os.path.join(val_lbl, label)
    )

print("Validation set created.")
```

```python
import os

def remap_yolo_labels(input_file, output_file, mapping):
    """
    Reads a YOLO label file, remaps class IDs according to 'mapping',
    and writes the updated labels to a new file.
    """

    new_lines = []

    with open(input_file, "r") as f:
        for line in f:
            line = line.strip()

            if not line:
                continue

            parts = line.split()

            cls = int(parts[0])

            # Keep only the classes we want
            if cls not in mapping:
                continue

            new_cls = mapping[cls]

            parts[0] = str(new_cls)

            new_lines.append(" ".join(parts))

    with open(output_file, "w") as f:
        f.write("\n".join(new_lines))
```

```python
import tempfile
import os

sample_input = tempfile.NamedTemporaryFile(delete=False, mode="w", suffix=".txt")
sample_input.write("2 0.5 0.5 0.2 0.2\n")
sample_input.write("0 0.4 0.4 0.1 0.1\n")
sample_input.close()

sample_output = sample_input.name.replace(".txt", "_out.txt")

HELMET_MAP = {
    2: 0,
    0: 1,
    1: 2,
    3: 3
}

remap_yolo_labels(sample_input.name, sample_output, HELMET_MAP)

with open(sample_output) as f:
    print(f.read())
```
